### Data cleaning

```python
df.head(3)
```

```python
### Strip Units & Convert to Numeric
def cleanunits(df):
    df["Mileage"]=df["Mileage"].str.split(" ").str[0].astype(float)
    df["Engine"]=df["Engine"].str.split(" ").str[0].astype(float)
    df["Power"] = df["Power"].str.split(" ").str[0]
    df["Power"] = pd.to_numeric(df["Power"], errors="coerce")
    return df
df=cleanunits(df)
```

```python
print(df.isnull().sum())
```

```python
def clean_data(df):
    df.drop("New_Price", axis=1)
    df["Mileage"].fillna(df["Mileage"].median())
    df["Engine"].fillna(df["Engine"].median())
    df["Power"].fillna(df["Power"].median())
    df["Seats"].fillna(df["Seats"].median())
    return df
df=clean_data(df)

print(df.isnull().sum())
```

```python
df['Brand']=df['Name'].str.split(" ").str[0]
df.drop("Name", axis=1, inplace=True)
```
